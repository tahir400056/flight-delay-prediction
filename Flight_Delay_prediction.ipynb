```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# ✅ Pipeline for numeric columns
numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),  # fill missing values
    ('scaler', StandardScaler())                   # scale values
])

# Fit + transform
X_numeric_prepared = numeric_pipeline.fit_transform(numeric)

print("✅ Numeric pipeline applied (imputation + scaling)")
print("Output shape:", X_numeric_prepared.shape)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, classification_report

# Columns split
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = X_train.select_dtypes(include=['object']).columns

# Pipelines
numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('scaler', StandardScaler())
])

categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="most_frequent")),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Column transformer
preprocessor = ColumnTransformer([
    ('num', numeric_pipeline, numeric_cols),
    ('cat', categorical_pipeline, categorical_cols)
])

# Final pipeline with Linear Regression
linreg_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LinearRegression())
])

# Train
linreg_pipeline.fit(X_train, y_train)

# Predict continuous values
y_pred_cont = linreg_pipeline.predict(X_test)

# Convert into classification (threshold=0.5)
y_pred = (y_pred_cont >= 0.5).astype(int)

# Evaluate
print("Linear Regression (as classifier) Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
